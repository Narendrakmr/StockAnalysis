```python
df3
```

```python
df3['date']=pd.to_datetime(df3['date'])
```

```python
df3['Daily_Price_Change']=df3['close']-df3['open']
```

```python
df3.sample()
```

```python
figure=px.line(df3,x='date',y='Daily_Price_Change',title='Daily Price Change',width=800,height=400)
figure.show()
```

```python
#Daily percentage return for a Amazon stock
df3['Daily_%_change']=(df3['Daily_Price_Change']/df3['close'])*100
```

```python
df3.sample()
```

```python
figure=px.line(df3,x='date',y='Daily_%_change',title='Daily Percent Change',width=800,height=400)
figure.show()
```
